```python
# data_preprocessing.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.preprocessing import StandardScaler
import math
import json
from tqdm.notebook import tqdm

# Set display options
pd.set_option('display.max_columns', None)
plt.style.use('ggplot')
%matplotlib inline

# Step 1: Load Ball-by-Ball Data
print("Loading data...")
deliveries_df = pd.read_csv('../data/raw/deliveries.csv')
matches_df = pd.read_csv('../data/raw/Cricket_data.csv')

print(f"Deliveries dataset shape: {deliveries_df.shape}")
print(f"Matches dataset shape: {matches_df.shape}")

# Display sample data
print("\nDeliveries sample:")
display(deliveries_df.head())

print("\nMatches sample:")
display(matches_df.head())

# Create output directory for processed data
os.makedirs('../data/processed/match_sequences', exist_ok=True)

# Step 3: Process ball-by-ball data into over-by-over sequences
def preprocess_match_data(deliveries, matches):
    """
    Process ball-by-ball deliveries into over-by-over sequences
    with additional contextual features.
    """
    # Rename the id column in matches to match_id for consistency
    matches = matches.rename(columns={'id': 'match_id'})
    
    # Ensure data types match for joining
    deliveries['match_id'] = deliveries['match_id'].astype(str)
    matches['match_id'] = matches['match_id'].astype(str)
    
    # Merge match information with deliveries
    try:
        merged_df = pd.merge(
            deliveries, 
            matches[['match_id', 'venue_name', 'start_date', 'home_team', 'away_team', 
                    'toss_won', 'decision', 'winner']], 
            on='match_id', 
            how='left'
        )
        print(f"Merged data shape: {merged_df.shape}")
    except Exception as e:
        print(f"Error merging data: {e}")
        print("Proceeding with deliveries data only")
        merged_df = deliveries.copy()
    
    # Process data into over-by-over format
    over_data = []
    
    # Group by match and innings
    for (match_id, inning), group in tqdm(merged_df.groupby(['match_id', 'inning'])):
        # Sort by over and ball
        group = group.sort_values(['over', 'ball'])
        
        # Get batting and bowling teams
        batting_team = group['batting_team'].iloc[0]
        bowling_team = group['bowling_team'].iloc[0]
        
        # Process each over
        for over, over_group in group.groupby('over'):
            # Calculate over statistics
            runs_in_over = over_group['total_runs'].sum()
            
            # Count wickets in this over
            wickets_in_over = over_group['is_wicket'].sum()
            
            # Calculate cumulative stats
            prev_overs = group[group['over'] < over]
            cumulative_runs = prev_overs['total_runs'].sum() + runs_in_over
            cumulative_wickets = prev_overs['is_wicket'].sum() + wickets_in_over
            
            # Calculate run rate
            current_over = over + 1  # Adjust for 0-indexed overs
            run_rate = cumulative_runs / current_over if current_over > 0 else 0
            
            # Calculate overs remaining (assuming T20 format - 20 overs)
            max_overs = 20  # Change to 50 for ODI
            overs_remaining = max_overs - current_over
            
            # Get current batsmen and bowler
            current_batsmen = list(set(
                list(over_group['batter'].unique()) + 
                list(over_group['non_striker'].unique())
            ))
            current_bowler = over_group['bowler'].iloc[-1]
            
            # Determine match phase
            if current_over <= 6:
                match_phase = 'Powerplay'
            elif current_over <= 15:
                match_phase = 'Middle'
            else:
                match_phase = 'Death'
            
            # Get venue information
            try:
                venue = group['venue_name'].iloc[0]
            except:
                venue = "Unknown"
            
            # Add to over data
            over_data.append({
                'match_id': match_id,
                'inning': inning,
                'over': over,
                'batting_team': batting_team,
                'bowling_team': bowling_team,
                'runs_in_over': runs_in_over,
                'wickets_in_over': wickets_in_over,
                'cumulative_runs': cumulative_runs,
                'cumulative_wickets': cumulative_wickets,
                'overs_remaining': overs_remaining,
                'run_rate': run_rate,
                'current_batsmen': current_batsmen,
                'current_bowler': current_bowler,
                'match_phase': match_phase,
                'venue': venue
            })
    
    # Convert to DataFrame
    over_df = pd.DataFrame(over_data)
    print(f"Created over-level dataset with {over_df.shape[0]} rows")
    
    return over_df

# Calculate momentum based on recent performance
def calculate_momentum(over_df, window_size=3):
    """
    Calculate momentum based on recent performance.
    """
    # Create a copy to avoid modifying the original
    df = over_df.copy()
    df['momentum'] = 0.0  # Default value
    
    # Process each match innings separately
    for (match_id, inning), group in tqdm(df.groupby(['match_id', 'inning'])):
        group = group.sort_values('over')
        
        # Calculate momentum for each over
        for i in range(len(group)):
            if i < window_size:
                # Not enough data for momentum calculation
                continue
                
            # Get recent overs
            recent_window = group.iloc[i-window_size:i]
            overall = group.iloc[:i]
            
            # Recent run rate vs overall run rate
            recent_rr = recent_window['runs_in_over'].mean()
            overall_rr = overall['runs_in_over'].mean()
            
            # Wicket impact (negative momentum)
            wicket_impact = -0.5 * recent_window['wickets_in_over'].sum()
            
            # Combined momentum index
            momentum = (recent_rr - overall_rr) + wicket_impact
            
            # Update momentum value in the main dataframe
            idx = group.index[i]
            df.loc[idx, 'momentum'] = momentum
    
    return df

# Create batsman statistics (strike rate, etc.)
def add_batsman_stats(over_df, deliveries_df):
    """
    Add batsman statistics like strike rate to the over-level data.
    """
    # Create a copy to avoid modifying the original
    df = over_df.copy()
    df['batsman_sr'] = 0.0  # Default value
    
    # Calculate batsman strike rates from ball-by-ball data
    batsman_stats = {}
    
    # Process each ball to collect batsman data
    for _, row in tqdm(deliveries_df.iterrows(), total=len(deliveries_df), desc="Calculating batsman stats"):
        batter = row['batter']
        runs = row['batsman_runs']
        
        if batter not in batsman_stats:
            batsman_stats[batter] = {'runs': 0, 'balls': 0}
        
        batsman_stats[batter]['runs'] += runs
        batsman_stats[batter]['balls'] += 1
    
    # Calculate strike rates
    for batter in batsman_stats:
        if batsman_stats[batter]['balls'] > 0:
            batsman_stats[batter]['sr'] = (batsman_stats[batter]['runs'] / batsman_stats[batter]['balls']) * 100
        else:
            batsman_stats[batter]['sr'] = 0
    
    # Add batsman strike rates to over data
    for idx, row in df.iterrows():
        batsmen = row['current_batsmen']
        if batsmen:
            # Calculate average strike rate of current batsmen
            sr_values = []
            for batsman in batsmen:
                if batsman in batsman_stats and 'sr' in batsman_stats[batsman]:
                    sr_values.append(batsman_stats[batsman]['sr'])
            
            if sr_values:
                df.loc[idx, 'batsman_sr'] = sum(sr_values) / len(sr_values)
    
    return df

# Create bowler statistics (economy rate, etc.)
def add_bowler_stats(over_df, deliveries_df):
    """
    Add bowler statistics like economy rate to the over-level data.
    """
    # Create a copy to avoid modifying the original
    df = over_df.copy()
    df['bowler_economy'] = 0.0  # Default value
    
    # Calculate bowler economy rates from over-level data
    bowler_stats = {}
    
    # First, group deliveries by match, bowler, and over
    bowler_overs = deliveries_df.groupby(['match_id', 'bowler', 'over']).agg({
        'total_runs': 'sum'
    }).reset_index()
    
    # Calculate economy rates
    for _, row in tqdm(bowler_overs.iterrows(), total=len(bowler_overs), desc="Calculating bowler stats"):
        bowler = row['bowler']
        runs = row['total_runs']
        
        if bowler not in bowler_stats:
            bowler_stats[bowler] = {'runs': 0, 'overs': 0}
        
        bowler_stats[bowler]['runs'] += runs
        bowler_stats[bowler]['overs'] += 1
    
    # Calculate economy rates
    for bowler in bowler_stats:
        if bowler_stats[bowler]['overs'] > 0:
            bowler_stats[bowler]['economy'] = bowler_stats[bowler]['runs'] / bowler_stats[bowler]['overs']
        else:
            bowler_stats[bowler]['economy'] = 0
    
    # Add bowler economy rates to over data
    for idx, row in df.iterrows():
        bowler = row['current_bowler']
        if bowler in bowler_stats and 'economy' in bowler_stats[bowler]:
            df.loc[idx, 'bowler_economy'] = bowler_stats[bowler]['economy']
    
    return df

# Step 4: Function to export each match as a sequence
def export_match_sequences(over_df, output_dir):
    """
    Export each match innings as a CSV sequence file.
    """
    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)
    
    # Create one-hot encoding for match phase
    phase_dummies = pd.get_dummies(over_df['match_phase'], prefix='phase')
    over_df = pd.concat([over_df, phase_dummies], axis=1)
    
    # Process each match innings
    sequence_count = 0
    for (match_id, inning), group in over_df.groupby(['match_id', 'inning']):
        # Skip incomplete innings (less than 5 overs)
        if len(group) < 5:
            continue
            
        # Sort by over
        group = group.sort_values('over')
        
        # Select columns for sequence data
        sequence_cols = [
            'over', 'runs_in_over', 'wickets_in_over', 'cumulative_runs', 
            'cumulative_wickets', 'overs_remaining', 'run_rate', 'momentum',
            'batsman_sr', 'bowler_economy', 'phase_Powerplay', 'phase_Middle', 
            'phase_Death'
        ]
        
        # Ensure all columns exist
        for col in ['phase_Powerplay', 'phase_Middle', 'phase_Death']:
            if col not in over_df.columns:
                sequence_cols.remove(col)
        
        # Create sequence dataframe
        sequence_df = group[sequence_cols].copy()
        
        # Add match metadata
        meta_data = {
            'match_id': str(match_id),
            'inning': int(inning),
            'batting_team': group['batting_team'].iloc[0],
            'bowling_team': group['bowling_team'].iloc[0],
            'venue': group['venue'].iloc[0],
            'final_score': int(group['cumulative_runs'].max()),
            'wickets_lost': int(group['cumulative_wickets'].max())
        }
        
        # Save as CSV
        output_file = f"{output_dir}/match_{match_id}_innings_{inning}.csv"
        sequence_df.to_csv(output_file, index=False)
        
        # Also save metadata as JSON
        meta_file = f"{output_dir}/match_{match_id}_innings_{inning}_meta.json"
        with open(meta_file, 'w') as f:
            json.dump(meta_data, f)
        
        sequence_count += 1
    
    print(f"Exported {sequence_count} match sequences to {output_dir}")

# Main execution
print("Processing ball-by-ball data into over-by-over sequences...")
match_data = preprocess_match_data(deliveries_df, matches_df)

print("Calculating momentum...")
match_data = calculate_momentum(match_data)

print("Adding batsman statistics...")
match_data = add_batsman_stats(match_data, deliveries_df)

print("Adding bowler statistics...")
match_data = add_bowler_stats(match_data, deliveries_df)

print("Exporting match sequences...")
export_match_sequences(match_data, '../data/processed/match_sequences')

print("Data preprocessing complete! Ready for model development.")

# Display sample of processed data
display(match_data.head())

# Visualize some key metrics
plt.figure(figsize=(12, 6))
sns.histplot(match_data['runs_in_over'], kde=True)
plt.title('Distribution of Runs per Over')
plt.xlabel('Runs')
plt.show()

plt.figure(figsize=(12, 6))
sns.lineplot(x='over', y='run_rate', data=match_data)
plt.title('Run Rate Progression by Over')
plt.xlabel('Over Number')
plt.ylabel('Run Rate')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='match_phase', y='runs_in_over', data=match_data)
plt.title('Runs Distribution by Match Phase')
plt.show()

# Analyze momentum
plt.figure(figsize=(12, 6))
sns.histplot(match_data['momentum'].dropna(), kde=True)
plt.title('Distribution of Momentum Values')
plt.xlabel('Momentum')
plt.show()

# Show relationship between momentum and runs in next over
match_data['next_over_runs'] = match_data.groupby(['match_id', 'inning'])['runs_in_over'].shift(-1)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='momentum', y='next_over_runs', data=match_data.dropna())
plt.title('Relationship Between Momentum and Runs in Next Over')
plt.xlabel('Momentum')
plt.ylabel('Runs in Next Over')
plt.show()
```
